```python
# sub-dataset para os tweets levando em considera√ß√£o a coluna 'created_at'
dfjb_1 = dfjb_clean[['created_at']]
dfjb_1 = dfjb_1.copy()

# Divis√£o da coluna 'created_at' em horas e dias
dfjb_1['hour'] = dfjb_1.created_at.dt.hour
dfjb_1['day'] = dfjb_1.created_at.dt.weekday

# Quantidade de tweets em cada hora do dia 0(Segundas)

dfjb_1.groupby('day').hour.value_counts()[0]
```

#### Para ser poss√≠vel comparar a atividade no Twitter durante cada hora de cada dia, a atividade foi definida como a porcentagem de postagens no Twitter durante o per√≠odo de 1 hora em dado dia da semana (quantidade de tweets postados a cada hora de um dia / quantidade total de tweets postados no mesmo dia). Para isso criaremos uma coluna chamada tweet_percentage.

```python
# porcentagem de tweets a cada hora de um dia da semana
dfjb_1 = dfjb_1.groupby('day').hour.value_counts() / dfjb_1.groupby('day').hour.agg('count') * 100
dfjb_1 = pd.DataFrame(dfjb_1).rename(columns = {'hour':'tweet_percentage'})

# reset multi-index
dfjb_1.reset_index(level = ['day','hour'], inplace = True)

display(dfjb_1)
```

##### Devido √† durante v√°rias horas de cada dia n√£o ter havido nenhuma postagem no twitter, essas horas exatas n√£o foram contabilizadas no dataframe dfjb_1

```python
#Para ser possivel plotar as 24h do dia de cada dia,adicionaremos essas horas sem publica√ß√£o ao dataframe.

ind = dfjb_1.index.max() + 1

# horas existentes: 0, 1, ..., 22, 23
hours = [hour for hour in range(24)]

# se a hora em quest√£o n√£o estiver no dataframe
for d in dfjb_1.day.unique():
    for hr in hours:
        if hr not in dfjb_1.query('day == @d').hour.unique():
            
            # adicionado nova linha com o dia, hora e 0% de postagem de twitter
            dfjb_1.loc[ind] = d, hr, 0
            
            # Atualizando o √≠ndice para a adi√ß√£o de linha seguinte
            ind += 1

display(dfjb_1)
```

```python
# sort values by three columns and reset index
dfjb_1.sort_values(by = ['day', 'hour', 'tweet_percentage'], inplace = True)
dfjb_1.reset_index(drop = True, inplace = True)

display(dfjb_1)
```

```python
day_of_week = ['Monday', 'Tuesday', 'Wednesday', 'Thursday', 'Friday', 'Saturday', 'Sunday']
dow = [day for day in np.arange(7)] # integers representing days of a week

# dictionary mapping integers to days of a week
dow_dict = {}
for key in dow:
    # integer: day of week (ex 0: Monday)
    dow_dict[key] = day_of_week[key]
```

```python
# Criando uma figura com 2x4 sub-plotagens
linhas = 2
colunas = 4
fig, ax = plt.subplots(nrows = linhas, ncols = colunas, figsize = (18, 9))

fig.suptitle('Porcentagens dos tweets a cada hora para cada dia da semana', fontsize = 18)
fig.text(x = 0.5, y = 0.06, s = 'Hora do dia', ha = 'center', va = 'center', fontsize = 16)
fig.text(x = 0.09, y = 0.5, s = 'Porcentagem de tweets postados (%)', ha = 'center', va = 'center', rotation = 'vertical', fontsize = 16)

# Para cada sub-plotagem, ax[linhas, colunas]
for row in np.arange(linhas):
    for col in np.arange(colunas):
        
        if linhas == 0: # in first row, col number = integer for the given day
            dfjb_1_temp = dfjb_1.query('day == @col') # filter data for the given day
            day_name = dow_dict[col] # find name of the day
            
        else:
            # delete last (eighth) subplot because there is only seven days in a week
            if col == colunas - 1:
                fig.delaxes(ax[row, col])
                break
            
            # in second row, col number = integer for the given day - 4
            dfjb_1_temp = dfjb_1.query('day == @col + 4')
            day_name = dow_dict[col + 4]

        # set up xtick locations and labels
        width = 2
        x_loc = list(dfjb_1_temp.hour * width)
        labels = [label for label in np.arange(24)]
        
        # bar plots for percentage of hourly tweets for each day of week
        ax[row, col].bar(x = x_loc, height = dfjb_1_temp.tweet_percentage, width = width, align = 'edge')
        
        # label suplot with name of the day
        ax[row, col].set_title(day_name, fontsize = 14)
        
        # label xticks by increments of two: 0, 2, ..., 22
        for i in range(len(labels)):
            if labels[i] % 2 != 0:
                labels[i] = ''
        

        # add xtick labels to the subplot
        ax[row, col].set_xticks(ticks = x_loc)
        ax[row, col].set_xticklabels(labels = labels)        
      
```

# Fun√ß√£o de Processamento

```python
def processamento(texto):
    texto=texto.lower()
    documento=pln(texto)
    lista=[]
    for p in documento:
        #lista.append(p)
        lista.append(p.lemma_) #Processo de lemmatiza√ß√£o
    lista2=[]
    for p in lista:
        if (p in stop_words)==False:
            lista2.append(p)
    lista3=[]
    for p in lista2:
        if (p in pontuacoes)==False:
            lista3.append(p)
    return lista3
```
